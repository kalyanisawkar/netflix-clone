### Bidirectional Associative Memory (BAM)
BAM is a hetero-associative neural network that associates patterns from one set (X) to patterns in another set (Y). It uses a weight matrix $W$ calculated as:
$W = \sum (X_i^T \cdot Y_i)$

```python
# Practical 5
# Bidirectional Associative Memory (BAM)

import numpy as np

# Input vector pairs
X1 = np.array([1, -1, 1])
Y1 = np.array([1, -1])

X2 = np.array([-1, 1, -1])
Y2 = np.array([-1, 1])

# Weight matrix calculation
W = np.outer(X1, Y1) + np.outer(X2, Y2)

print("Weight Matrix:")
print(W)

# Test input
X = np.array([1, -1, 1])

# Forward association
Y = np.dot(X, W)

# Activation function
Y = np.where(Y >= 0, 1, -1)

print("\nAssociated Output:")
print(Y)

# Backward association
X_back = np.dot(Y, W.T)

# Activation function
X_back = np.where(X_back >= 0, 1, -1)

print("\nRecovered Input:")
print(X_back)
```

```python
#LAB ASSIGNMENT 6

import numpy as np

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Input dataset (XOR function)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Initialize weights randomly with mean 0
np.random.seed(1)
weights_input_hidden = np.random.uniform(size=(2, 2))  # Weights from input to hidden layer
weights_hidden_output = np.random.uniform(size=(2, 1))  # Weights from hidden to output layer

# Bias terms
bias_hidden = np.random.uniform(size=(1, 2))  # Bias for hidden layer
bias_output = np.random.uniform(size=(1, 1))  # Bias for output layer

# Training parameters
learning_rate = 0.1
epochs = 10000

# Training loop
for epoch in range(epochs):
    # Forward pass
    hidden_layer_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_layer_output = sigmoid(hidden_layer_input)

    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
    predicted_output = sigmoid(output_layer_input)

    # Backpropagation
    # Calculate error at output
    error = y - predicted_output
    d_predicted_output = error * sigmoid_derivative(predicted_output)

    # Calculate error at hidden layer
    error_hidden = d_predicted_output.dot(weights_hidden_output.T)
    d_hidden_layer = error_hidden * sigmoid_derivative(hidden_layer_output)

    # Update weights and biases
    weights_hidden_output += hidden_layer_output.T.dot(d_predicted_output) * learning_rate
    bias_output += np.sum(d_predicted_output, axis=0, keepdims=True) * learning_rate

    weights_input_hidden += X.T.dot(d_hidden_layer) * learning_rate
    bias_hidden += np.sum(d_hidden_layer, axis=0, keepdims=True) * learning_rate

# Print final predictions
print("Final Predictions:")
hidden_layer_input = np.dot(X, weights_input_hidden) + bias_hidden
hidden_layer_output = sigmoid(hidden_layer_input)
output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
predicted_output = sigmoid(output_layer_input)

for i in range(len(X)):
    print(f"Input: {X[i]}, Predicted Output: {predicted_output[i][0]:.4f}, Expected Output: {y[i][0]}")
```

```python
import numpy as np

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, learning_rate=0.1):
        # 1. Initialize weights and biases with random values
        self.weights_input_hidden = np.random.uniform(-1, 1, (input_size, hidden_size))
        self.weights_hidden_output = np.random.uniform(-1, 1, (hidden_size, output_size))
        self.bias_hidden = np.random.uniform(-1, 1, (1, hidden_size))
        self.bias_output = np.random.uniform(-1, 1, (1, output_size))
        self.lr = learning_rate

    # Sigmoid activation function
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    # Derivative of sigmoid for backpropagation
    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def train(self, inputs, targets, epochs):
        for epoch in range(epochs):
            # --- PHASE 1: Forward Propagation ---
            hidden_layer_input = np.dot(inputs, self.weights_input_hidden) + self.bias_hidden
            hidden_layer_output = self.sigmoid(hidden_layer_input)

            output_layer_input = np.dot(hidden_layer_output, self.weights_hidden_output) + self.bias_output
            predicted_output = self.sigmoid(output_layer_input)

            # --- PHASE 2: Backpropagation ---
            # Calculate output layer error and gradient
            output_error = targets - predicted_output
            d_output = output_error * self.sigmoid_derivative(predicted_output)

            # Calculate hidden layer error and gradient
            hidden_error = d_output.dot(self.weights_hidden_output.T)
            d_hidden = hidden_error * self.sigmoid_derivative(hidden_layer_output)

            # --- PHASE 3: Weight and Bias Updates ---
            self.weights_hidden_output += hidden_layer_output.T.dot(d_output) * self.lr
            self.bias_output += np.sum(d_output, axis=0, keepdims=True) * self.lr
            self.weights_input_hidden += inputs.T.dot(d_hidden) * self.lr
            self.bias_hidden += np.sum(d_hidden, axis=0, keepdims=True) * self.lr

    def predict(self, inputs):
        hidden = self.sigmoid(np.dot(inputs, self.weights_input_hidden) + self.bias_hidden)
        output = self.sigmoid(np.dot(hidden, self.weights_hidden_output) + self.bias_output)
        return output

# Example Usage: Training on XOR logic
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

nn = NeuralNetwork(input_size=2, hidden_size=4, output_size=1, learning_rate=0.2)
nn.train(X, y, epochs=20000)

print("Predictions after 20,000 epochs:")
for val in X:
    print(f"Input: {val} Output: {nn.predict(val)}")
```

```python
import numpy as np

class HopfieldNetwork:
    def __init__(self, size):
        self.size = size
        self.weights = np.zeros((size, size))

    def train(self, patterns):
        # Hebbian Learning Rule
        for p in patterns:
            self.weights += np.outer(p, p)
        # Set diagonal to 0 (no self-connections)
        np.fill_diagonal(self.weights, 0)
        self.weights /= self.size

    def predict(self, pattern, steps=5):
        res = np.copy(pattern)
        for _ in range(steps):
            for i in range(self.size):
                # Asynchronous update
                raw_input = np.dot(self.weights[i], res)
                res[i] = 1 if raw_input >= 0 else -1
        return res

# Define 4 patterns to store (8-bit vectors for simplicity)
p1 = np.array([1, 1, 1, 1, -1, -1, -1, -1])
p2 = np.array([1, -1, 1, -1, 1, -1, 1, -1])
p3 = np.array([1, 1, -1, -1, 1, 1, -1, -1])
p4 = np.array([-1, 1, -1, 1, -1, 1, -1, 1])

patterns = [p1, p2, p3, p4]

# Initialize and train network
hn = HopfieldNetwork(size=8)
hn.train(patterns)

# Test retrieval with a noisy version of p1
noisy_p1 = np.array([1, 1, 1, 1, -1, -1, 1, 1]) # Last two bits flipped
recovered_p1 = hn.predict(noisy_p1)

print("Original P1: ", p1)
print("Noisy P1:    ", noisy_p1)
print("Recovered:   ", recovered_p1)

# Verify if recovered matches original
if np.array_equal(p1, recovered_p1):
    print("\nSuccess: The network recovered the original pattern!")
else:
    print("\nNote: The network converged to a different stable state or local minimum.")
```

```python
# CNN using TensorFlow with Graph Plotting

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Reshape data
X_train = X_train.reshape(60000, 28, 28, 1) / 255.0
X_test = X_test.reshape(10000, 28, 28, 1) / 255.0

# Create CNN model
model = Sequential()

# Convolution Layer
model.add(Conv2D(
    32,
    (3,3),
    activation='relu',
    input_shape=(28,28,1)
))

# Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Flatten Layer
model.add(Flatten())

# Fully Connected Layer
model.add(Dense(128, activation='relu'))

# Output Layer
model.add(Dense(10, activation='softmax'))

# Compile Model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train Model
history = model.fit(
    X_train,
    y_train,
    epochs=3,
    validation_data=(X_test, y_test)
)

# Evaluate Model
loss, accuracy = model.evaluate(X_test, y_test)

print("Accuracy :", accuracy)

# Plot Accuracy Graph
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.legend(['Training Accuracy', 'Validation Accuracy'])

plt.show()

# Plot Loss Graph
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.legend(['Training Loss', 'Validation Loss'])

plt.show()
```

### MNIST Digit Classification using PyTorch
This implementation uses the PyTorch framework to load the MNIST dataset, define a Convolutional Neural Network, and evaluate its accuracy.

```python
#4. MNIST Handwritten Character Detection using PyTorch, Keras and Tensorflow

import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# 1. Hyperparameters and Data Loading
batch_size = 64
learning_rate = 0.01
epochs = 3

transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# 2. Define CNN Architecture
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(32 * 13 * 13, 128)
        self.fc2 = nn.Linear(128, 10)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.pool(x)
        x = x.view(-1, 32 * 13 * 13)
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = SimpleCNN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# 3. Training Loop
model.train()
for epoch in range(epochs):
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
    print(f'Epoch {epoch+1}/{epochs} completed.')

# 4. Evaluation
model.eval()
correct = 0
with torch.no_grad():
    for data, target in test_loader:
        output = model(data)
        pred = output.argmax(dim=1, keepdim=True)
        correct += pred.eq(target.view_as(pred)).sum().item()

print(f'\nPyTorch Test Accuracy: {100. * correct / len(test_loader.dataset):.2f}%')
```
